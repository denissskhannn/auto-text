**Задание 5. Лемматизация и стемминг**

Примените к токенам алгоритмы лемматизации и стемминга. Выведите 2 списка - 1. Лемматизированные токены 2. Стемматизированные токены

```python
def detect_language(word):
    if re.search('[а-яё]', word.lower()):
        return 'ru'
    elif re.search('[a-z]', word.lower()):
        return 'en'
    return 'unknown'
```

```python
import subprocess
subprocess.run(["python", "-m", "spacy", "download", "ru_core_news_sm"])
subprocess.run(["python", "-m", "spacy", "download", "en_core_web_sm"])
nlp_ru = spacy.load("ru_core_news_sm")
nlp_en = spacy.load("en_core_web_sm")
def lemmatize_with_spacy(tokens):
    lemmatized = []

    russian_tokens = []
    english_tokens = []

    for token in tokens:
        lang = detect_language(token)
        if lang == 'ru':
            russian_tokens.append(token)
        elif lang == 'en':
            english_tokens.append(token)
        else:
            lemmatized.append(token)

    if russian_tokens:
        russian_text = ' '.join(russian_tokens)
        doc_ru = nlp_ru(russian_text)
        for token in doc_ru:
            if token.text.strip():
                lemmatized.append(token.lemma_)

    if english_tokens:
        english_text = ' '.join(english_tokens)
        doc_en = nlp_en(english_text)
        for token in doc_en:
            if token.text.strip():
                lemmatized.append(token.lemma_)

    return lemmatized

lemmatized_tokens = lemmatize_with_spacy(cleaned_tokens)

russian_stemmer = SnowballStemmer("russian")
english_stemmer = SnowballStemmer("english")

stemmed_tokens = []
for token in cleaned_tokens:
    lang = detect_language(token)
    if lang == 'ru':
        stemmed_tokens.append(russian_stemmer.stem(token))
    elif lang == 'en':
        stemmed_tokens.append(english_stemmer.stem(token))
    else:
        stemmed_tokens.append(token)

print("Лемматизированные токены:")
print(lemmatized_tokens)
print()
print("Стемматизированные токены:")
print(stemmed_tokens)
print()
```

**Задание 6. Напишите функцию для препроцессинга текста**

Объедините все шаги в одну функцию. Выведите результат с лемматизированным списком

```python

def preprocess_text(text, method='lemmatization'):

    text = text.lower()
    text = text.replace('ё', 'е')
    text = text.replace('\n', ' ')

    punctuation = string.punctuation + '—°⋅' + '«»„\"'
    text = text.translate(str.maketrans('', '', punctuation))

    text = re.sub(r'\d+', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    russian_stopwords = set(stopwords.words('russian'))

    cleaned_tokens = []
    for token in tokens:
        if token not in russian_stopwords:
            cleaned_tokens.append(token)

    if method == 'lemmatization':
        russian_tokens = []
        english_tokens = []

        for token in cleaned_tokens:
            lang = detect_language(token)
            if lang == 'ru':
                russian_tokens.append(token)
            elif lang == 'en':
                english_tokens.append(token)

        result = []

        if russian_tokens:
            russian_text = ' '.join(russian_tokens)
            doc_ru = nlp_ru(russian_text)
            for token in doc_ru:
                if token.text.strip():
                    result.append(token.lemma_)

        if english_tokens:
            english_text = ' '.join(english_tokens)
            doc_en = nlp_en(english_text)
            for token in doc_en:
                if token.text.strip():
                    result.append(token.lemma_)

    elif method == 'stemming':
        result = []
        for token in cleaned_tokens:
            lang = detect_language(token)
            if lang == 'ru':
                result.append(russian_stemmer.stem(token))
            elif lang == 'en':
                result.append(english_stemmer.stem(token))
            else:
                result.append(token)
    else:
        result = cleaned_tokens

    return result

result = preprocess_text(text, method='lemmatization')
print(result)
```
