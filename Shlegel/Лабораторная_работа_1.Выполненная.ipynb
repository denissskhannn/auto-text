**Задание 1. Загрузка данных**

Вставьте текст для обработки согласно вашему варианту

```python
text = """В 2023 году нейроинтерфейс Neuralink успешно имплантирован 12 пациентам с тетраплегией, позволив им управлять компьютером силой мысли со скоростью до 25 слов в минуту.
Технология BCI (Brain-Computer Interface) использует массив из 1024 электродов толщиной 5 мкм для считывания нейронной активности.
"Мы стоим на пороге телепатической эры," — заявил известный нейрофизиолог на конференции NeuroTech-2024.
Китайский проект BrainNet достиг скорости передачи данных 104 бит/с между мозгом и компьютером, что в 8,7 раза быстрее предыдущего рекорда!
По данным исследования, проведенного в 7 странах, 58% респондентов в возрасте 18-35 лет готовы использовать неинвазивные нейроинтерфейсы для повседневной коммуникации.
Этические проблемы mind-reading технологий вызывают острые дискуссии.
ООН в 2023 году приняла резолюцию №42/157 "О защите нейроприватности", запрещающую несанкционированный доступ к мыслям человека.
Эксперты прогнозируют, что к 2035 году около 15% населения Земли будет использовать ту или иную форму нейроинтерфейса."""
```

**Задание 2. Нормализация текста.**

Приведите текст к нижнему регистру, удалите лишние пробоелы, переносы строк, спецсимволы, пунктуацию, обработайте цифры.

```python
# ваш код
def normalize_text(text):
  text = text.lower()
  text = text.replace('\n', ' ')
  text = re.sub(r'\d+', '', text)
  text = re.sub(r'[^\w\s]', ' ', text)
  text = ' '.join(text.split())
  return text

normalized_text = normalize_text(text)
print(normalized_text)
```

**Задание 3. Токенизация**

Токенизируйте текст.

```python
# ваш код
tokens = word_tokenize(normalized_text)
print(tokens)
```

**Задание 4. Удаление стоп-слов**

Выведите 2 списка - 1. Очищенных токенов, 2. Список удаленных стоп-слов

```python
# ваш код
russian_stopwords = set(stopwords.words('russian'))

clean_tokens = [token for token in tokens if token not in russian_stopwords]
removed_words = [token for token in tokens if token in russian_stopwords]

print(clean_tokens)
print(removed_words)
```

**Задание 5. Лемматизация и стемминг**

Примените к токенам алгоритмы лемматизации и стемминга. Выведите 2 списка - 1. Лемматизированные токены 2. Стемматизированные токены

```python
# ваш код
nlp_ru = spacy.load("ru_core_news_sm")
def lemmas(tokens):
  joined = ' '.join(tokens)
  doc = nlp_ru(joined)
  return [token.lemma_ for token in doc if token.is_alpha]

ru_lemmas = lemmas(clean_tokens)

stemmer = SnowballStemmer("russian")
stemmed_tokens = [stemmer.stem(token) for token in clean_tokens]

print(ru_lemmas)
print(stemmed_tokens)
```

**Задание 6. Напишите функцию для препроцессинга текста**

Объедините все шаги в одну функцию. Выведите результат с лемматизированным списком

```python
# ваш код

def preprocess(text):
  text = text.lower()
  text = text.replace('\n', ' ')
  text = re.sub(r'\d+', '', text)
  text = re.sub(r'[^\w\s]', ' ', text)
  text = ' '.join(text.split())

  tokens = word_tokenize(text)

  russian_stopwords = set(stopwords.words('russian'))

  clean_tokens = [token for token in tokens if token not in russian_stopwords]
  removed_words = [token for token in tokens if token in russian_stopwords]

  nlp_ru = spacy.load("ru_core_news_sm")
  joined = ' '.join(clean_tokens)
  doc = nlp_ru(joined)
  ru_lemmas = [token.lemma_ for token in doc if token.is_alpha]

  stemmer = SnowballStemmer("russian")
  stemmed_tokens = [stemmer.stem(token) for token in clean_tokens]

  return ru_lemmas

result = preprocess(text)
print(result)
```
