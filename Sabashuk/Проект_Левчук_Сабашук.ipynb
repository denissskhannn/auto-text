```python
# Загрузка библиотек
!pip install nltk

import os
import re
from collections import Counter
import nltk
from nltk import word_tokenize, pos_tag
from nltk.corpus import stopwords
import matplotlib.pyplot as plt

nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download("stopwords")
```

```python
# Чтение текста из файла
with open("guardian_iwd_corpus.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()

print("Символов в корпусе:", len(raw_text))
```

```python
# Предобработка текста
STOPWORDS = set(stopwords.words("english"))

KEYWORDS = {
    "woman", "women",
    "lady", "ladies",
    "girl", "girls",
    "female", "females"
}

ADJ_TAGS = {"JJ", "JJR", "JJS"}
VERB_TAGS = {"VB", "VBD", "VBG", "VBN", "VBP", "VBZ"}
MODAL_TAGS = {"MD"}
BE_FORMS = {"be", "am", "is", "are", "was", "were", "been", "being"}

CORPUS_PATH = "guardian_iwd_corpus.txt"
#OUTPUT_PATH = "gender_pos_patterns.txt"
#os.makedirs(OUTPUT_PATH, exist_ok=True)

PATTERN_TYPES = {
    "ADJ + KEYWORD": Counter(),
    "ADJ + ADJ + KEYWORD": Counter(),
    "KEYWORD + VERB": Counter(),
    "KEYWORD + MODAL + VERB": Counter(),
    "KEYWORD + WORD + VERB": Counter(),
}

def preprocess_and_tag(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    text = re.sub(r"\s+", " ", text)

    tokens = word_tokenize(text)
    tokens = [t for t in tokens if t not in STOPWORDS]

    return pos_tag(tokens)

tagged_tokens = preprocess_and_tag(raw_text)
print("Токенов после очистки и pos-теггинга:", len(tagged_tokens))
```

```python
# Извлечение паттернов
def extract_patterns(tagged_tokens):

    for i in range(len(tagged_tokens)):
        word, tag = tagged_tokens[i]

        # 1. ADJ + KEYWORD
        if i > 0:
            w1, t1 = tagged_tokens[i - 1]
            if word in KEYWORDS and t1 in ADJ_TAGS:
                PATTERN_TYPES["ADJ + KEYWORD"][(w1, word)] += 1

        # 2. ADJ + ADJ + KEYWORD
        if i > 1:
            w1, t1 = tagged_tokens[i - 1]
            w2, t2 = tagged_tokens[i - 2]
            if word in KEYWORDS and t1 in ADJ_TAGS and t2 in ADJ_TAGS:
                PATTERN_TYPES["ADJ + ADJ + KEYWORD"][(w2, w1, word)] += 1

        # 3. KEYWORD + VERB
        if word in KEYWORDS and i + 1 < len(tagged_tokens):
            w1, t1 = tagged_tokens[i + 1]
            if t1 in VERB_TAGS:
                PATTERN_TYPES["KEYWORD + VERB"][(word, w1)] += 1

        # 4. KEYWORD + MODAL + VERB
        if word in KEYWORDS and i + 2 < len(tagged_tokens):
            w1, t1 = tagged_tokens[i + 1]
            w2, t2 = tagged_tokens[i + 2]
            if t1 in MODAL_TAGS and t2 in VERB_TAGS:
                PATTERN_TYPES["KEYWORD + MODAL + VERB"][(word, w1, w2)] += 1

        # 5. KEYWORD + WORD + VERB
        if word in KEYWORDS and i + 2 < len(tagged_tokens):
            w1, _ = tagged_tokens[i + 1]
            w2, t2 = tagged_tokens[i + 2]
            if t2 in VERB_TAGS:
                PATTERN_TYPES["KEYWORD + WORD + VERB"][(word, w1, w2)] += 1

extract_patterns(tagged_tokens)
```

```python
# Вывод топ-20 паттернов
for pattern_type, counter in PATTERN_TYPES.items():
    print(f"\nTOP-10: {pattern_type}\n")
    for pat, freq in counter.most_common(10):
        print(f"{' '.join(pat)} → {freq}")
```

```python
# Сохранение результатов в отдельный файл
OUTPUT_FILE = "all_gender_patterns.txt"

with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    for pattern_type, counter in PATTERN_TYPES.items():

        # Заголовок секции
        f.write("=" * 60 + "\n")
        f.write(pattern_type.upper() + "\n")
        f.write("=" * 60 + "\n")

        # Паттерны и частотность
        for pat, freq in counter.most_common():
            f.write(f"{' '.join(pat)}\t{freq}\n")

        f.write("\n\n")  # пустая строка между секциями

print(f"Все паттерны сохранены в файл: {OUTPUT_FILE}")
```

```python
# Визуализация
def plot_patterns(counter, title, top_k=10):
    top = counter.most_common(top_k)
    labels = [" ".join(p) for p, _ in top]
    values = [v for _, v in top]

    plt.figure(figsize=(10, 6))
    plt.barh(
        labels[::-1],
        values[::-1],
        color="#d8b4e2"  # светло-сиреневый
    )
    plt.title(title)
    plt.xlabel("Frequency")
    plt.tight_layout()
    plt.show()

for pattern_type, counter in PATTERN_TYPES.items():
    plot_patterns(counter, f"Top-10 patterns: {pattern_type}")
```
