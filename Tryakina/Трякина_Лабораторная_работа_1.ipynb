**Задание 0. Загрузите необходимые библиотеки**

```python
!pip install spacy
!python -m spacy download en_core_web_sm # англ словарь
!python -m spacy download ru_core_news_sm # рус словарь

import spacy
import pandas as pd
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stopWords = stopwords.words('russian')

nlp_en = spacy.load('en_core_web_sm')
nlp_ru = spacy.load('ru_core_news_sm')

from nltk.stem import SnowballStemmer

english_stemmer = SnowballStemmer("english")
russian_stemmer = SnowballStemmer("russian")
```

**Задание 1. Загрузка данных**

Вставьте текст для обработки согласно вашему варианту

```python
text = """Синтетическая биология — междисциплинарная область, объединяющая инженерный подход с генетикой.
В 2024 году методы CRISPR/Cas9 позволяют редактировать геном с точностью до 1 нуклеотида!
Компания Ginkgo Bioworks создала дрожжи, производящие 16 различных ароматических соединений, что удешевило парфюмерное производство на 68%.
"Bio-engineering становится новой промышленной революцией XXI века," — считает профессор Ли из MIT.
Микроорганизмы, выращенные в 5000-литровых биореакторах, синтезируют материалы, идентичные шёлку и кожи, без участия животных.
В 2023 году стартап NatureLoop представил biodegradable пластик на основе целлюлозы, который разлагается в почве за 90 дней.
Искусственные бактерии способны превращать отходы агропромышленности в биотопливо с эффективностью до 72%!
Согласно прогнозам, к 2030 году до 35% химической промышленности перейдет на bio-based производство.
Этические вопросы регулирования новой отрасли обсуждаются в рамках международного соглашения Biological Code Act (BCA-2025)."""
```

**Задание 2. Нормализация текста.**

Приведите текст к нижнему регистру, удалите лишние пробоелы, переносы строк, спецсимволы, пунктуацию, обработайте цифры.

```python
# ваш код
def norm_text(text):
  text = text.lower()
  text = text.replace('%', ' процент')
  text = text.replace('\n', ' ') # конец строки на пробел
  text = re.sub(r'[^\w\s]', ' ', text) # убираем пунктуацию и спецсимволы
  text = re.sub(r'\s+', ' ', text).strip()
  return text

work_text = norm_text(text)
print(work_text)
```

**Задание 3. Токенизация**

Токенизируйте текст.

```python
# ваш код
def tok_text(text):
  words = re.findall(r'\b\w+\-?\w+\b|\b\w+\b', text)
  return words

work_words = tok_text(work_text)
print(work_words)
```

**Задание 4. Удаление стоп-слов**

Выведите 2 списка - 1. Очищенных токенов, 2. Список удаленных стоп-слов

```python
# ваш код
def del_stops(words, stopwords):
  clean = []
  stops = []
  for word in words:
    if word in stopwords:
      stops.append(word)
    else:
      clean.append(word)
  return [clean, stops]

work_words, stop_words = del_stops(work_words, stopWords)
print('Очищенные токены: ', work_words)
print('Список удаленных стоп-слов', stop_words)
```

**Задание 5. Лемматизация и стемминг**

Примените к токенам алгоритмы лемматизации и стемминга. Выведите 2 списка - 1. Лемматизированные токены 2. Стемматизированные токены

```python
# ваш код
def stem_words(words): # word[0].isascii() типа проверка на лат алф
  return [english_stemmer.stem(word) if word[0].isascii() else russian_stemmer.stem(word) for word in words]

def lem_words(words):
  lemmy = []
  for word in words:
    if word[0].isascii():
      lemmy.append(nlp_en(word)[0].lemma_)
    else:
      lemmy.append(nlp_ru(word)[0].lemma_)
  return lemmy

lemmy = lem_words(work_words)
stemmy = stem_words(work_words)
print('Лемматизированные токены: ', lemmy)
print('Стемматизированные токены: ', stemmy)
```

**Задание 6. Напишите функцию для препроцессинга текста**

Объедините все шаги в одну функцию. Выведите результат с лемматизированным списком

```python
# ваш код
def do_it(text):
  work_words, stop_words = del_stops(tok_text(norm_text(text)), stopWords)
  data1 = {'Очищенные слова' : work_words}
  data2 = {'Стоп слова' : stop_words}
  panda1 = pd.DataFrame(data1)
  panda2 = pd.DataFrame(data2)
  display(panda1)
  display(panda2)
  stemmy = stem_words(work_words)
  lemmy = lem_words(work_words)
  data3 = {
      'Слово' : work_words,
      'Стемма' : stemmy,
      'Лемма' : lemmy
  }
  panda3 = pd.DataFrame(data3)
  display(panda3)
  return lemmy

print(do_it(text))
```
