**Задание 2. Нормализация текста.**

Приведите текст к нижнему регистру, удалите лишние пробоелы, переносы строк, спецсимволы, пунктуацию, обработайте цифры.

```python
def normalize_text(text):
    text = text.lower()
    text = ''.join(char for char in text if not char.isdigit())
    translator = str.maketrans(string.punctuation, ' ' * len(string.punctuation))
    text = text.translate(translator)
    text = ' '.join(text.split())
    return text

normalized_text = normalize_text(text)
print(normalized_text)
```

**Задание 3. Токенизация**

Токенизируйте текст.

```python
def tokenize(text): 
    text = normalize_text(text) 
    tokens = text.split() 
    return tokens

tokens = tokenize(text)
print(tokens)
```

**Задание 4. Удаление стоп-слов**

Выведите 2 списка - 1. Очищенных токенов, 2. Список удаленных стоп-слов

```python
stop_words = set(stopwords.words('russian'))

def rem_stopwords(words, stopwords):
    clean = []
    removed = []

    for word in words:
        if word in stop_words:
            removed.append(word)
        else:
            clean.append(word)

    return [clean, removed]

tokens, removed_stopwords = rem_stopwords(tokens, stopwords)
print("Очищенные токены: ", tokens)
print("Список удаленных стоп-слов", removed_stopwords)
```

**Задание 5. Лемматизация и стемминг**

Примените к токенам алгоритмы лемматизации и стемминга. Выведите 2 списка - 1. Лемматизированные токены 2. Стемматизированные токены

```python
import spacy
nlp_en = spacy.load('en_core_web_sm')
nlp_ru = spacy.load('ru_core_news_sm')

def lm_words(words):
    lemmed = []
    for word in words:
        if word[0].isascii():
            lemmed.append(nlp_en(word)[0].lemma_)
        else:
            lemmed.append(nlp_ru(word)[0].lemma_)
    return lemmed

def s_words(words):
    stemmed = []
    for word in words:
        if word[0].isascii():
            stemmed.append(english_stemmer.stem(word))
        else:
            stemmed.append(russian_stemmer.stem(word))
    return stemmed

lemmy = lm_words(tokens)
stemmy = s_words(tokens)
print("Лемматизированные токены: ", lemmy)
print("Стемматизированные токены: ", stemmy)
```
