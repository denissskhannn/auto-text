**Задание 1. Загрузка данных**

Вставьте текст для обработки согласно вашему варианту

```python
text = "В 2024 году искусственный интеллект (AI) совершил революцию в медицинской диагностике. Нейросети типа MedVision-3 способны обнаруживать раковые клетки с точностью до 97,8%, превосходя опытных онкологов на 12,3%. Группа исследователей из MIT и Сколтеха разработала алгоритм DeepDiagnosis, который анализирует МРТ-снимки в 45 раз быстрее человека! 'Мы наблюдаем начало новой эры в здравоохранении,' — заявил профессор Иванов на конференции AI-Med 2024. Однако сохраняются этические вопросы: кто несет ответственность за ошибку ИИ-диагноста? Согласно опросу, только 62% пациентов доверяют AI-заключениям без подтверждения врача. Интеграция machine learning в больницах требует пересмотра медицинских протоколов и подготовки специалистов, способных эффективно взаимодействовать с 'умными' помощниками. В России к 2025 году планируется внедрить AI-системы в 75% федеральных медцентров."
```

**Задание 2. Нормализация текста.**

Приведите текст к нижнему регистру, удалите лишние пробоелы, переносы строк, спецсимволы, пунктуацию, обработайте цифры.

```python
def normalize_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\d+', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

normalized_text = normalize_text(text)
print(normalized_text)
```

**Задание 3. Токенизация**

Токенизируйте текст.

```python
tokens = word_tokenize(normalized_text)
print(f"Количество токенов: {len(tokens)}")
print(f"Первые 10 токенов: {tokens[:10]}")
```

**Задание 4. Удаление стоп-слов**

Выведите 2 списка - 1. Очищенных токенов, 2. Список удаленных стоп-слов

```python
stop_words = set(stopwords.words('russian'))

filtered_tokens = [t for t in tokens if t not in stop_words]

removed_stop_words = [t for t in tokens if t in stop_words]

print("Очищенные токены:", filtered_tokens[:])
print("Удаленные стоп-слова:", list(set(removed_stop_words)))
```

**Задание 5. Лемматизация и стемминг**

Примените к токенам алгоритмы лемматизации и стемминга. Выведите 2 списка - 1. Лемматизированные токены 2. Стемматизированные токены

```python
morph = pymorphy3.MorphAnalyzer()
stemmer = SnowballStemmer("russian")

lemmatized_tokens = [morph.parse(t)[0].normal_form for t in filtered_tokens]

stemmed_tokens = [stemmer.stem(t) for t in filtered_tokens]

print("Леммы:", lemmatized_tokens[:])
print("Стеммы:", stemmed_tokens[:])
```

**Задание 6. Напишите функцию для препроцессинга текста**

Объедините все шаги в одну функцию. Выведите результат с лемматизированным списком

```python
def full_preprocess(input_text):
    clean_text = normalize_text(input_text)

    raw_tokens = word_tokenize(clean_text)

    russian_stops = set(stopwords.words('russian'))
    no_stops = [t for t in raw_tokens if t not in russian_stops]

    morph_analyzer = pymorphy3.MorphAnalyzer()
    final_lemmas = [morph_analyzer.parse(t)[0].normal_form for t in no_stops if len(t) > 2]

    return final_lemmas

result = full_preprocess(text)
print("Итоговый список лемм:")
print(result)
```
